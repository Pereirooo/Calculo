```python
# a. Construye g(t) = f2(f1(t)). Obtén su dominio

import sympy as sp

t = sp.Symbol('t', real=True)       #Primero definimos las variables que vamos a utilizar
expr1 = sp.cos(t)                   #Realizamos la expresión 1
expr2 = sp.sqrt(t)                  #Realizamos la expresión 2
f1 = sp.Lambda(t, expr1)            # Empleamos sp.Lambda para crear la función f1 a partir de l expresión 1
f2 = sp.Lambda(t, expr2)            # Empleamos otra vez sp.Lambda para crear la función f2 a partir de la expresión 2
display (f1(t))                     # Comprobamos que esté bien 
display (f2(t))                     # También comprobamos 
expr_g = f2(f1(t))                  # Realizamos la expresión a construir 
g = sp.Lambda(t, expr_g)            # Definimos la función requerida en el enunciado.
display(g(t))


#Procedemos a obtener el dominio

display(sp.calculus.singularities(g(t), t))     # Para calcular el dominio de dicha función procedemos calculando primero las singularidades de la función y el dominio será todo R menos las singularidades de la función.

# En este caso al obtener 0, esto nos indica que el dominio de g(t) es todos los números reales menos el 0.
```

```python
# b. Estudia las simetrías de f1

# Para estudiar las simetrías de la función f1 estudiaremos primero la par y luego la impar 
# Para que una función tenga simetría par: f(x)=f(-x)
# Para analizar esto vamos a emplear lo que llamamos una hipótesis sobre una variable, realizamos una serie de comparaciones/condiciones y dependiendo si se cumplen o no, tendremos un resultado u otro.

expr1 = f1(t)
expr2 = f1(-t)
solucion1 = sp.solve(expr1,t)
solucion2 = sp.solve(expr2,t)
solucion1 == solucion2
# Al devolvernos True podemos afirmar que la función "f1" tiene simetría par ya que cumple la condición, de todas formas comprobamos si en la hipótesis de la impar nos devuleve False.
# Para que una función tenga simetría impar: -f(x)=f(-x)

expr3 = -f1(t)
expr4 = f1(-t)
solucion3 = sp.solve(expr3,t)
solucion4 = sp.solve(expr4,t)
solucion3 == solucion4
```
