```python
# log((1+x)/(1-x))

#)Primera parte. Dominio de la función
import sympy as sp
import numpy as np

x = sp.symbols('x', real=True) # define la variable simbólica x

f = sp.log((1+x)/(1-x))
f_expr = sp.Lambda(x,f)
display(f)
I = sp.calculus.util.continuous_domain(f, x, sp.Reals)
display(sp.calculus.singularities(f, x))
print('El dominio de la función es:',I)

p = sp.plot((f),(x,-20,20), show=False) 
p[0].line_color='r'
p.xlabel='x'
p.ylabel='y'
p.legend=True
p.show()


# display(sp.solve(f_expr(0),x))
```

```python
#)Primera parte. Concavidad y convexidad de la función
import sympy as sp
import numpy as np

x = sp.symbols('x', real=True) # define la variable simbólica x

f = sp.log((1+x)/(1-x))
f_expr = sp.Lambda(x,f)
display(f)
 

f_der=  sp.diff(f,x,1)
f_der2=  sp.diff(f,x,2)
print('La primera derivada es:',f_der)
print('La segunda derivada es:',f_der2)


pc = sp.solve(f_der)
print('Los puntos críticos son', pc)
pi = sp.solve(f_der2)
print('Los puntos de inflexión son', pi)
```

```python
#a)Segunda parte. Polinomio de MacLaurin de grado 2
display('Apartado a')
import numpy as np
import sympy as sp
import matplotlib as mp
import matplotlib.pyplot as plt

from math import factorial

%matplotlib inline

x= sp.symbols('x', real=True) # define la variable simbólica x

# definimos la función
f_expr = sp.log((1+x)/(1-x))
f = sp.Lambda(x,f_expr)

print('El valor desp.log((1+0)/(1-0)) es: ', f(0),'\n')

# definimos la recta tangente por 0
f_der=  sp.diff(f(x),x)
f1 = sp.Lambda(x,f_der)
rt = f1(0)*(x-0) + f(0)

print('La recta tangente a f(x) =',f(x),' en x=0 es: P1(x) =', rt,'\n')

# punto de interés
x0 = 0

# Creamos gráficos de funciones
x1 = np.linspace(1, 10, 200)
y1 = np.log(x1)
y2 = f1(0)*(x1-0) + f(0)

fig, axs = plt.subplots(1, 2, figsize=(20,10))

ax1 = axs[0]
ax1.plot(x1, y1, c='b', lw='3',  label='$f(x)=\ln(x)$')
ax1.plot(x1, y2, c='k', ls='--', lw='3', label='$P_1(x)=x-1$')
ax1.axvline(x=1.3, c='red', ls='-.', lw='2')
ax1.set_ylabel('Y', fontsize=10)
ax1.set_xlabel('X', fontsize=10)
ax1.grid()
ax1.legend(prop={'size': 18})


ax2 = axs[1]
ax2.plot(x1, y1, c='b', lw='3', label='$f(x)=\ln(x)$')
ax2.plot(x1, y2, c='k', ls='--', lw='3', label='$P_1(x)=x-1$')
ax2.axvline(x=1.3, c='red', ls='-.', lw='2')
ax2.set_ylabel('Y', fontsize=10)
ax2.set_xlabel('X', fontsize=10)
plt.xlim(0.8,1.4)
plt.ylim(-0.3,0.5)
ax2.set_xticks(np.arange(0.8,1.5,0.1))
ax2.set_yticks(np.arange(-0.3,0.6,0.1))
ax2.grid()
ax2.legend(prop={'size': 18})



x0 = 0 # punto en el que centramos el polinomio de Taylor

# función que queremos aproximar
fx = sp.log((1+x)/(1-x))

# calculamos el Polinomio de Taylor de grado n centrado en x0
n = 1  # grado del polinomio
P1 = sp.series(fx,x,x0,2)
display(P1) 

# la O() representa los términos de orden superior que aparecerían en un polinomio de orden superior, por lo que la eliminamos
P1=P1.removeO()

# que coincide con la recta tangente calculada anteriormente (rt)
display(P1==rt)

# calculamos el polinomio de Taylor de orden 2
P2 = sp.series(fx,x,x0,3)
P2=P2.removeO()
print('El polinomio de Taylor de orden 2 es:')
display(P2)


# Creamos gráficos de funciones
x1 = np.linspace(0.4, 2.5, 200)
y1 = np.log(x1)
y2 = f1(1)*(x1-1) + f(1)
# evaluamos P2 en los puntos de x1
p2x = np.zeros(len(x1))
for k in range(len(x1)):
    p2x[k] = P2.subs({x:x1[k]})
    

fig, axs = plt.subplots(1, 2, figsize=(20,10))

ax1 = axs[0]
ax1.plot(x1, y1, c='b', lw='2',  label='$f(x)=\ln(x)$')
ax1.plot(x1, y2, c='k', ls='--', lw='2', label='$P_1(x)=x-1$')
ax1.plot(x1, p2x, c='r', ls='--', lw='2', label='$P_2(x)=x-1 - \dfrac{(x-1)^2}{2}$')
ax1.set_ylabel('Y', fontsize=10)
ax1.set_xlabel('X', fontsize=10)
ax1.grid()
ax1.legend(prop={'size': 18})


ax2 = axs[1]
ax2.plot(x1, y1, c='b', lw='2', label='$f(x)=\ln(x)$')
ax2.plot(x1, y2, c='k', ls='--', lw='2', label='$P_1(x)=x-1$')
ax2.plot(x1, p2x, c='r', ls='--', lw='2', label='$P_2(x)=x-1 - \dfrac{(x-1)^2}{2}$')
ax2.set_ylabel('Y', fontsize=10)
ax2.set_xlabel('X', fontsize=10)
plt.xlim(0.8,1.4)
plt.ylim(-0.3,0.5)
ax2.set_xticks(np.arange(0.8,1.5,0.1))
ax2.set_yticks(np.arange(-0.3,0.6,0.1))
ax2.grid()
ax2.legend(prop={'size': 18})





#b)Aproximación de log(3) y acotación del error cometido en la aproximación
display('Apartado b')
#Comprobación del valor de log(3)

print('El valor de log(3) es',sp.log(3))
print('El valor de 2*log(3) utilizando el polinomio de Taylor es', 2*sp.log(3))


#error

x = sp.symbols('x', real=True)

f= sp.log((1+x)/(1-x))

f_der1=  sp.diff(f,x)
f_der2=  sp.diff(f_der1,x)
f_der3=  sp.diff(f_der2,x)

ferr = sp.Lambda(x,f_der2)
display(ferr)

fsup = sp.Abs(ferr(0.5))

error1 = (fsup*((0.5-0)**2))/(factorial(2)) #está elevado a 2 y es factorial de 2 porque es n+1 y p es de grado 1

print('El error tiene que ser menor o igual que', error1)


```
